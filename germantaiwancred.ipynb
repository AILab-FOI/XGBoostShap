```python
import os
import pandas as pd
import numpy as np

dataset_name = "taiwan"
data_path = "data/taiwan_credit_default.csv"

if not os.path.exists(data_path):
    raise FileNotFoundError(f"Datoteka nije pronađena: {data_path}")

df = pd.read_csv(data_path)

print("Dimenzije dataseta:", df.shape)
display(df.head())
df.info()

missing = df.isnull().sum()
print("\nNedostajuće vrijednosti po stupcima:")
display(missing[missing > 0])

print("\nTipovi podataka:")
display(df.dtypes.value_counts())

num_cols = df.select_dtypes(include=[np.number]).columns
display(df[num_cols].describe().T)

duplicates = df.duplicated().sum()
print("\nBroj dupliciranih redaka:", duplicates)

id_cols = [c for c in df.columns if c.lower() in ["id", "client_id", "customer_id"]]
for col in id_cols:
    print(f"Duplikati po stupcu {col}:", df[col].duplicated().sum())

possible_targets = [
    "default payment next month",
    "default_payment_next_month",
    "DEFAULT",
    "target",
    "label"
]

target_col = None
for c in possible_targets:
    if c in df.columns:
        target_col = c
        break

if target_col:
    print("\nCiljna varijabla:", target_col)
    display(df[target_col].value_counts())
    display(df[target_col].value_counts(normalize=True))
else:
    print("\nCiljna varijabla nije automatski prepoznata.")
```
